```python
import galsim
import numpy as np
import matplotlib.pyplot as plt
```

```python
lam = 692

aberrations = np.zeros(12)
aberrations[4] = 1. #defocus
aberrations[5] = 0.35 #astig1
aberrations[6] = -0.25 #astig2
aberrations[7] = -0.35 #coma1
aberrations[8] = -0.5 #coma2
aberrations[9] = 0.25 #tref1
aberrations[10] = -0.5 #tref2
aberrations[11] = 0.25 #spherical

n = 15
shapes = np.zeros((n,3))
for i in range(n):
    optics = galsim.PhaseScreenList(galsim.AtmosphericScreen(r0_500=.15, L0=75, altitude=0.5, 
                                                             screen_size=25, vx=2, vy=1.2),
                                    galsim.AtmosphericScreen(r0_500=.12, L0=75, altitude=20, 
                                                             screen_size=25, vx=3.4, vy=4.2),
                                    galsim.OpticalScreen(diam=8.1, obscuration=1.024/8.1, aberrations=aberrations))
    aper = galsim.Aperture(diam=8.1, lam=lam, obscuration=1.024/7.7, screen_list=optics)
    psfIm = galsim.PhaseScreenPSF(optics, lam, exptime=15, aper=aper, flux=10**4).drawImage(scale=.01, 
                                                                                            nx=256, ny=256)
    moments = galsim.hsm.FindAdaptiveMom(psfIm)
    shapes[i] = [moments.observed_shape.g1, moments.observed_shape.g2, moments.moments_sigma]
```

```python
plt.figure(figsize=(8,4))
plt.subplot(121)
plt.hist(shapes[:,0], histtype='step', label='g1')
plt.hist(shapes[:,1], histtype='step', label='g2')
plt.xlabel('ellipticity')
plt.legend()

plt.subplot(122)
plt.hist(shapes[:,2]*.011*2.355, histtype='step')
plt.xlabel('size [pixels]')
plt.show()
```
