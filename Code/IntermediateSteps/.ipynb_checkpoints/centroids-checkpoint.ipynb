### Try fitting a slope to centroid motion

```python
centroidSlopeDictA = {}

for file in fileNames[:int(len(fileNames)/2)]:
    pts = np.linspace(0,999,N)
    x = centroidDictA[file.split('.')[0].split('_')[-1]]['x']
    y = centroidDictA[file.split('.')[0].split('_')[-1]]['y']
    
    px = np.polyfit(pts, x, deg=1)
    py = np.polyfit(pts, y, deg=1)
    
    centroidSlopeDictA[file] = {'x': px[0]*200, 'y': py[0]*200}
    
centroidSlopeDictB = {}

for file in fileNames[int(len(fileNames)/2):]:
    pts = np.linspace(0,999,N)
    x = centroidDictB[file.split('.')[0].split('_')[-1]]['x']
    y = centroidDictB[file.split('.')[0].split('_')[-1]]['y']
    
    px = np.polyfit(pts, x, deg=1)
    py = np.polyfit(pts, y, deg=1)
    
    centroidSlopeDictB[file] = {'x': px[0]*200, 'y': py[0]*200}
```

```python
plt.figure(figsize=(8,5))
plt.plot(pts, y, 'o', ms=3, color='lightblue')
plt.plot(pts, py[0] * pts + py[1], color='orange')
plt.xticks([0,250,500,750,1000],[0,15,30,45,60])
plt.ylabel('PSF y position [pixel]', fontsize=12)
plt.xlabel('exposure time', fontsize=12)
# plt.savefig('../../Plots/exampleCentroidSlope.png')
plt.show()
```

```python
xs = []
ys = []

for file in centroidSlopeDictB.keys():
    xs.append(centroidSlopeDictB[file]['x'])
    ys.append(centroidSlopeDictB[file]['y'])
    
for file in centroidSlopeDictA.keys():
    xs.append(centroidSlopeDictA[file]['x'])
    ys.append(centroidSlopeDictA[file]['y'])
    
plt.figure(figsize=(9,5))
plt.hist(xs, bins=25, histtype='step',label='x',color='turquoise',linestyle='--',linewidth=2)
plt.hist(ys, bins=25, histtype='step',label='y',color='gold',linewidth=1)
plt.xlabel('slope [pixel/minute]', fontsize=12)
plt.legend(fontsize=12)
# plt.savefig('../../Plots/centroidDistribution.png')
plt.show()
```
